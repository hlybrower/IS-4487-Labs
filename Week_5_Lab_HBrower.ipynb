```python
## Add at least two steps to filter the dataset
# remove any rows that don't have a source 
df.dropna(subset=['source'], inplace=True)

# remove any duplicates in the data
df.drop_duplicates()


## Add at least two charts to show a summary of any major trends
# scatter plot showing if there's any association between replies and retweets
df.plot.scatter(x='retweet_count',y='reply_count')

# favorites by whether truncated (longer than 140 characters)
by_truncated = df.groupby('truncated')['favorite_count'].sum()
by_truncated.plot.barh(color='orange', xticks=[0,5,10,15,20,25,30], align = 'edge', figsize = (6,6), fontsize = 8)







```
