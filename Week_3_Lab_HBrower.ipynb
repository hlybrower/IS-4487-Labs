## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

There aren't 5000 values for calls over 15  mins per month, and the minimum value for income and house amounts is negative which doesn't make sense in the context.

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]

## adding a filter to remove the rows that don't have a value for calls over 15 mins per month
df = df.dropna(subset =['over_15mins_calls_per_month'])

## make it so all values in the leave/stay category are uppercase, because some are lowercased which would make it difficult to view
df['leave'] = df['leave'].str.upper()

df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```
